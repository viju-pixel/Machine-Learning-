```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("D:\\datasets\\suv_car_data.csv")
data.head()
```

```python
real_x=data.iloc[:,[2,3]].values
real_x
```

```python
real_y=data.iloc[:,4].values
real_y
```

```python
# now training or testing data ko split krenge
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(real_x,real_y,test_size=0.25,random_state=0) 
```

```python
train_x # train the 75% data of 100% data
```

```python
test_x  # test the 25% data of 100% data
```

```python
 # now we required feature scaling  
from sklearn.preprocessing import StandardScaler    
```

```python
scaler=StandardScaler() # its provide the value b/w -2 to +2 range to reduce the gap b/w age and salary
```

```python
train_x=scaler.fit_transform(train_x)
train_x
```

```python
test_x=scaler.fit_transform(test_x)
test_x
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_Reg= LogisticRegression()
```

```python
Log_Reg.fit(train_x,train_y) #our  model is trained is here now let's move on prediction
```

```python
y_pred=Log_Reg.predict(test_x) # provide the testing data for testing 

y_pred  # this is my testing data after predicting 
```

```python
test_y # this is my testing data
```

```python
from sklearn.metrics import confusion_matrix
#it saperate the right and wrong prediction for matching the result and find the accurate result
```

```python
cm=confusion_matrix(test_y,y_pred) # here test_y is the actual value and y_pred is the prediction value 
cm
```

```python
# here 88 is the right prediction and 12 is wrong prediction so 88% is accuracy of my model
```

```python
 from matplotlib.colors import ListedColormap
```

```python
plt.contour(x1,x2,Log_Reg.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.75,
            cmap=ListedColormap(('red','green')))
            
```
